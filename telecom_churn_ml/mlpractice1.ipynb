```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub
import pandas as pd

path = kagglehub.dataset_download("blastchar/telco-customer-churn")

import os
print(os.listdir(path))  # see exact file name

df=pd.read_csv(path+"/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()



```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['gender'].unique()
```

```python
df['gender']=df['gender'].map({'Female':1, 'Male':0})
```

```python
pure_yes_no = [col for col in df.columns
               if set(df[col].dropna().unique()) == {'Yes', 'No'}]
```

```python
pure_yes_no
```

```python
df[pure_yes_no] = df[pure_yes_no].apply(lambda col: col.map({'Yes':1,'No':0}))
```

```python
df.head()
```

```python
df['MultipleLines'].unique()
```

```python
df['MultipleLines']=df['MultipleLines'].map({'Yes':1, 'No':0, 'No phone service':0})
```

```python
df['InternetService'].unique()
```

```python
new_cols = [col for col in df.columns
        if set(df[col].dropna().unique()) <= {'Yes', 'No', 'No internet service'}]
```

```python
new_cols
```

```python
df[new_cols] = df[new_cols].apply(lambda col: col.map({'Yes':1,'No':0,'No internet service':0}))
```

```python
df.info()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', None)
```

```python
df['TotalCharges']=df['TotalCharges'].astype(float)
```

```python
df.info()
```

```python
spe_cols=['InternetService','Contract','PaymentMethod']
df=pd.get_dummies(df,columns=spe_cols,drop_first=True)
df.info()
```

```python
df.columns
```

```python
df.info()
```

```python
df=df.drop('customerID',axis=1)
```

```python
df.head()
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.head()
```

```python
df['tenure'].unique()
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
df.isnull().sum()
```

```python
corr = df.corr()['Churn'].sort_values(ascending=False)
corr
```

```python
df = df.drop(['PhoneService', 'gender','MultipleLines'] , axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier


X_df=df.drop('Churn',axis=1)
y_df=df['Churn']
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X_df)

```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_df, test_size=0.2, random_state=42)
model_logistic=LogisticRegression()
model_logistic.fit(X_train,y_train)
y_predictLogistic=model_logistic.predict(X_test)
y_predictLogistic
```

```python
accuracy_logistic=model_logistic.score(X_test,y_test)
accuracy_logistic
```

```python
model_knn=KNeighborsClassifier()
model_knn.fit(X_train,y_train)
y_predictKNN=model_knn.predict(X_test)
accuracy_knn=model_knn.score(X_test,y_test)
accuracy_knn
```

```python
model_svm=SVC()
model_svm.fit(X_train,y_train)
y_predictSVM=model_svm.predict(X_test)
accuracy_svm=model_svm.score(X_test,y_test)
accuracy_svm
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)
model_tree=DecisionTreeClassifier()
model_tree.fit(X_train,y_train)
model_tree.predict(X_test)
accuracy_tree=model_tree.score(X_test,y_test)
accuracy_tree
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
scores_1=cross_val_score(model_logistic,X_scaled,y_df,cv=7)
scores_1.mean()
```

```python
scores_2=cross_val_score(model_svm,X_df, y_df, cv=7)
scores_2.mean()
```

```python
param={
    'C': [0.01, 0.1, 1, 10, 100, 1000],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga'],
    'max_iter': [500, 1000]
}
```

```python
grid_search = GridSearchCV(
    estimator=LogisticRegression(max_iter=1000, random_state=42),
    param_grid=param,           # ← param_grid is correct
    cv=7,
    scoring='accuracy',
    n_jobs=-1,                  # Use all cores - faster
    verbose=1                   # Shows progress
)

grid_search.fit(X_scaled, y_df)


```

```python
print("Best parameters: ", grid_search.best_params_)
print("Best accuracy: ", grid_search.best_score_)
```

```python
from sklearn.ensemble import RandomForestClassifier

final_rfc=RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    )
random_Score=cross_val_score(final_rfc,X_df,y_df,cv=7)
random_Score.mean()
```

```python
param_dist = {
    'n_estimators': [300, 400, 500],
    'max_depth': [10,12],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

rf=RandomForestClassifier(random_state=42)
rf_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=2,
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1,
    verbose=1
)

rf_search.fit(X_df, y_df)
```

```python
print("Best CV accuracy:", round(rf_search.best_score_, 4))
print("Best parameters:", rf_search.best_params_)
```
